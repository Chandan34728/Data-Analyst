# ➤ Count of Customers by Churn

```python
c= ["#06D001","#74E209"]
ax = sns.countplot(x = df["Churn"], data=df, palette=c)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Count of Cutomers by Churn")
plt.show()
```

#Based on given CoyntPlot we conclude that the number of churned customers is 1,869, while the number of non-churned customers is 5,174.

# ➤ Percentage of Churn Customer

```python
gp = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gp["Churn"], labels = gp.index, autopct = "%1.2f%%", colors = ["#9BEC00","#F3FF90"])
plt.title("Percentage of Churn Customer")
plt.show()
```

#From the given pie chart we can conclude that 26.54% of our customer have churned out.

# ➤ Churn of Customer by Gender

```python
ax = sns.countplot(x = df["gender"], data = df, hue = 'Churn', palette = 'BuGn_r')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churn by Gender")
plt.show()
```

#The churn distribution appears similar across genders, with both males and females showing comparable numbers of churned and non-churned customers. This suggests that gender does not significantly affect the likelihood of churn in this dataset..

# ➤ Count of Senior Citizen

```python
c = ['#06D001','#9BEC00']
ax = sns.countplot(x = df["SeniorCitizen"], data = df, palette = c)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title('Count of Senior Citizen')
plt.show()
```

#Based on the count plot above, we can conclude that the number of senior citizens is 1,142.

# ➤ Percentage of Churn by Senior Citizen

```python
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')

count_data['Percentage'] = count_data.groupby('SeniorCitizen')['Count'].transform(lambda x: x / x.sum() * 100)

pivot_data = count_data.pivot(index='SeniorCitizen', columns='Churn', values='Percentage')

ax = pivot_data.plot(kind='bar', stacked=True)

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

plt.title('Percentage of Churn by Senior Citizen Status')
plt.ylabel('Percentage')
plt.xlabel('Senior Citizen')
plt.show()
```

#From the plot, senior citizens have a significantly higher churn rate (41.7%) compared to non-senior citizens (23.6%). This indicates that senior citizens are more likely to churn than non-senior citizens.

# ➤ Churn of Customer Based on Services Used by Them

```python
# List of columns to create count plots for
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots with 3 rows and 3 columns
fig, axes = plt.subplots(3, 3, figsize=(15, 12))

# Flatten the axes for easy iteration
axes = axes.flatten()

# Loop through the columns and plot a count plot for each
c = ['#998CEB','#77E4D4']
for i, col in enumerate(columns):
    ax = sns.countplot(x=col, data=df, ax=axes[i], hue = "Churn", palette = c)
    axes[i].set_title(f'Count plot for {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])

# Adjust the layout
plt.tight_layout()
plt.show()
```

#The data reveals that customers who subscribe to additional services, such as **online security**, **online backup**, and **tech support**, tend to have lower churn rates compared to those who do not. In contrast, customers using **Fiber optic internet** show a higher tendency to churn compared to those with **DSL** or no internet service. **Phone service** customers also exhibit higher churn, though most customers still subscribe to it. The absence of **streaming services** like **StreamingTV** and **StreamingMovies** is also linked with higher churn. Overall, offering and encouraging subscription to these additional services could potentially reduce churn rates significantly.
